```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import io
%matplotlib inline
```

```python
data_clean=pd.read_csv('pisa2012_main.csv')
data_clean_top_bottom_country=data_clean[data_clean['country'].isin(["China (Shanghai) ",'Singapore','Hong Kong-China','Qatar','Peru ','Kazakhstan '])]
```

# Score differences between Gender, Country and Perceived Control

## What matters and what don't

## Countries differences

## Here are the differences in Science Score between all the countries.Â¶
China (Shanghai), Hong Kong-China, and Singapore are on the top. Qatar, Kazakhstan, and Peru on the bottom of this ranking.


```python
plt.figure(figsize=(12,12))
base_color=sb.color_palette()[0]
Science_score_means=data_clean.groupby('country')['Reading_score'].mean()
order=Science_score_means.sort_values(ascending=False).index
data_clean_math=data_clean['country'].value_counts()
sb.barplot(x="Reading_score", y="country", data=data_clean, order=order,color=base_color,ci=95)
plt.title('Reading Score: the difference between all the countries',fontsize=14)
```

## What are the difference between the top and the bottom countries

### China (Shanghai), Hong Kong - China, Singapore for the top - Peru, Qatar and Indonesia for the bottom

Boxplot show there are 75% Confidence Interval

```python
plt.figure(figsize = [20, 10])
base_color = sb.color_palette()[0]
order_country=['China (Shanghai) ','Hong Kong-China','Singapore','Kazakhstan ','Peru ','Qatar']
# left plot: violin plot
plt.subplot(1, 3, 1)
ax1 = sb.boxplot(data = data_clean_top_bottom_country, x = 'country', y = 'Math_score', order=order_country,color = base_color)
plt.title('Math Score')
plt.xticks(rotation=30)
# right plot: box plot
plt.subplot(1, 3, 2)
sb.boxplot(data = data_clean_top_bottom_country, x = 'country', y = 'Reading_score', order=order_country,color = base_color)
plt.title('Reading Score')
plt.xticks(rotation=30)
#plt.ylim(ax1.get_ylim()) # set y-axis limits to be same as left plot
plt.subplot(1, 3, 3)
sb.boxplot(data = data_clean_top_bottom_country, x = 'country', y = 'Science_score', order=order_country,color = base_color)
plt.xticks(rotation=30)
plt.title('Science Score')
plt.suptitle("Scores: big differences between China/Singapour and Peru/Qatar",fontsize=20)
```

## Score Difference: There isn't any depending on Gender

### Math, Reading and Science Score aren't so different between male and female

```python
ticks=np.arange(0,500,50)
plt.figure(figsize = [10, 5])
base_color = sb.color_palette()[0]
# left plot: violin plot
plt.subplot(1, 3, 1)
ax1 = sb.barplot(data = data_clean, x = 'Gender', y = 'Math_score', color = base_color)
plt.title('Math Score')
plt.yticks(ticks,ticks)
# right plot: box plot
plt.subplot(1, 3, 2)
sb.barplot(data = data_clean, x = 'Gender', y = 'Reading_score', color = base_color)
plt.title('Reading Score')
plt.yticks(ticks,ticks)
#plt.ylim(ax1.get_ylim()) # set y-axis limits to be same as left plot
plt.subplot(1, 3, 3)
sb.barplot(data = data_clean, x = 'Gender', y = 'Science_score', color = base_color)
plt.title('Science Score')
plt.yticks(ticks,ticks)
plt.suptitle('Scores: No difference between male and female',fontsize=15)
```

## Perceived Control: there is a slight difference

### This variable contained the most difference in my exploration

```python
ticks=np.arange(0,1000,100)
plt.figure(figsize = [20, 10])
base_color = sb.color_palette()[0]
order_demands_regardless=['Strongly disagree','Disagree','Agree','Strongly agree']
# left plot: violin plot
plt.subplot(1, 3, 1)
ax1 = sb.boxplot(data = data_clean, x = 'Perform_Poorly_Regardless', y = 'Math_score', order=order_demands_regardless,color = base_color)
plt.title('Math Score')
plt.yticks(ticks,ticks)
plt.xticks(rotation=30)
plt.xlabel('Perform Poorly Regardless')
# right plot: box plot
plt.subplot(1, 3, 2)
sb.boxplot(data = data_clean, x = 'Perform_Poorly_Regardless', y = 'Reading_score', order=order_demands_regardless,color = base_color)
plt.title('Reading Score')
plt.xticks(rotation=30)
plt.yticks(ticks,ticks)
plt.ylabel('Perform Poorly Regardless')
#plt.ylim(ax1.get_ylim()) # set y-axis limits to be same as left plot
plt.subplot(1, 3, 3)
sb.boxplot(data = data_clean, x = 'Perform_Poorly_Regardless', y = 'Science_score', order=order_demands_regardless,color = base_color)
plt.xticks(rotation=30)
plt.yticks(ticks,ticks)
plt.ylabel('Perform Poorly Regardless')
plt.title('Science Score')
plt.suptitle("Scores: Slight variation depending on Perform Poorly Regardless values",fontsize=20)
```

## Math, Reading and Science Scores depending on Countries and Gender

### Countries differences are still significant but don't create a difference depending on gender

```python
ticks=np.arange(0,1000,100)
plt.figure(figsize = [20, 10])
base_color = sb.color_palette()[0]
order_country=['China (Shanghai) ','Hong Kong-China','Singapore','Kazakhstan ','Peru ','Qatar']
order_mindset_perform_poorly=['Strongly disagree','Disagree','Agree','Strongly agree']

# left plot: violin plot
plt.subplot(1, 3, 1)
ax1 = sb.boxplot(data = data_clean_top_bottom_country, x = 'country', y = 'Math_score', hue = 'Gender',order=order_country
                  )
plt.title('Math Score')
plt.xticks(rotation=30)
plt.yticks(ticks,ticks)
plt.subplot(1, 3, 2)
ax1 = sb.boxplot(data = data_clean_top_bottom_country, x = 'country', y = 'Reading_score', hue = 'Gender',order=order_country,
                  )
plt.title('Reading Score')
plt.xticks(rotation=30)
plt.yticks(ticks,ticks)
plt.subplot(1, 3, 3)
ax1 = sb.boxplot(data = data_clean_top_bottom_country, x ='country' , y = 'Science_score', hue ='Gender' ,order=order_country)
plt.yticks(ticks,ticks)
plt.title('Science Score')
plt.xticks(rotation=30)
plt.suptitle('Scores: differences depending on countries, but not on gender',fontsize=20)
```
